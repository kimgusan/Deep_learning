# 1. 선형회귀 모델

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([2, 4, 6, 8])
y = np.array([81, 93, 91, 97])

plt.scatter(x, y)
plt.show()
```

```python
x = np.array([2, 4, 6, 8])
y = np.array([81, 93, 91, 97])

# 기울기 절편 초기화
a = 0 
b = 0

# 학습률 조정
lr = 0.03

# 반복 횟수 설정
epochs = 2001

# x 의 값이 총 몇개인지 확인
n = len(x)

# 경사하강법 실행
for i in range(epochs):
    y_pred = a * x + b
    # 실제 값과 비교한 오차를 error 로 설정
    error = y - y_pred

    # 오차 함수를 a로 편미분한 값입니다.
    a_diff = (2/n) * sum(-x * (error))
    # 오차 함수를 b로 편미분한 값입니다.
    b_diff = (2/n) * sum(-(error))

    # 학습률을 곱해서 기존의 a값을 업데이트 합니다.
    a = a -lr * a_diff
    # 학습률을 곱해서 기존의 b값을 업데이트 합니다.
    b = b -lr * b_diff

    if i % 100 ==0:
        print(f"epoch={i}, 기울기= {a:.4f}, 절편= {b:.4f}")

# 선 그래프 수식
y_pred = a * x + b

# 그래프 출력.
plt.scatter(x, y)
plt.plot(x, y_pred, 'r')
plt.grid()
plt.show()
```

# 3. tensorflow 선형 회귀

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# 텐서플로를 사용한 학습시간에 따른 점수 회귀 훈련 진행
x = np.array([2, 4, 6, 8])
y = np.array([81, 93, 91, 97])

model = Sequential()

model.add(Dense(1, input_dim = 1, activation = 'linear'))
model.compile(optimizer='sgd', loss ='mse')
model.fit(x, y, epochs=2000)

plt.scatter(x, y)
plt.plot(x, model.predict(x), 'r')
plt.show()
```

```python
# 임의의 시간을 집어넣어 점수를 예측하는 모델 사용
hour = 7
hour_array = np.array([hour])
prediction = model.predict(hour_array)
print(prediction)
print(f"{hour}시간을 공부할 경우의 예상 점수는 {prediction[0][0]:.2f}점 입니다.")
```

### 변수가 2개일 때 선형 회귀 분석

```python
# 텐서플로를 사용한 학습시간에 따른 점수 회귀 훈련 진행
x = np.array([[2, 0], [4, 4], [6, 2], [8, 3]])
y = np.array([81, 93, 91, 97])

model = Sequential()

model.add(Dense(1, input_dim = 2, activation = 'linear'))
model.compile(optimizer='sgd', loss ='mse')
model.fit(x, y, epochs=2000)
```

```python
# 임의의 시간을 집어넣어 점수를 예측하는 모델 사용
hour = 7
private_class = 4

input_data = np.array([[hour, private_class]])  # 두 입력값을 하나의 배열로 결합
prediction = model.predict(input_data)

print(f"{hour}시간을 공부하고 {private_class}시간의 과외를 받을 경우 예상 점수는 {prediction[0][0]: .2f}점 입니다.")
```

# 3. 로지스틱 회귀 모델

```python
# 텐서플로를 사용한 학습시간에 따른 점수 회귀 훈련 진행
x = np.array([2, 4, 6, 8, 10, 12, 14])
y = np.array([0, 0, 0, 1, 1, 1, 1])

model = Sequential()

model.add(Dense(1, input_dim = 1, activation = 'sigmoid'))
model.compile(optimizer='sgd', loss ='binary_crossentropy')
model.fit(x, y, epochs=5000)
```

```python
plt.scatter(x, y)
plt.plot(x, model.predict(x), 'r')
plt.show()
```

```python
# 임의의 시간을 집어넣어 점수를 예측하는 모델 사용
hour = 7

hour_array = np.array([hour])
prediction = model.predict(hour_array)

print(f"{hour}시간을 공부할 경우 합격 예상 확률은 {prediction[0][0] * 100: .2f}점 입니다.")
```
